In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])

df.head()
```

```python
# check the parameters for randomly generated data
df.apply(pd.DataFrame.describe, axis=1)
```

```python
# calculate the 95% confidence intervals for mean
stderr = scipy.stats.sem(df, axis = 1)
critical_value = scipy.stats.norm.ppf(1-(1-0.95)/2) # two-tailed (1.96)
delta = stderr * critical_value

# merge with data based on index
delta_df = pd.DataFrame(data = delta[:],
                        index = df.index.values,
                        columns = ['delta'])
```

```python
# user's input 
# y_value = 40000 # hard-coded value
try:
    y_value = (float
               (input
                ('Enter the value for y axis (ranging from {0} to {1}): '
                 .format(round(df.mean(axis = 1).min(), -4),
                         round(df.mean(axis = 1).max(), -4))  # round to closest ten-thousands
                )
               )
              )
    if y_value >= 30000 and y_value <= 50000:
        y_value
    elif y_value <= 0:
         y_value = round(df.median(axis = 1).median(), -1)
    else:      
        y_value = round(df.mean(axis = 1).min(), -1)
except:
    y_value = round(df.median(axis = 1).median(), -1)
    
```

```python
# final y_value used
print('Final value for y axis used:', y_value)
```

```python
# create temp dataframes to comply with the colouring conditions
below_y = df[round(df.mean(axis = 1),0) < y_value]
equal_to_y = df[round(df.mean(axis = 1),0) == y_value]
higher_than_y = df[round(df.mean(axis = 1),0) > y_value]
```

```python
# assign the value of delta to the temp dataframes 
delta_below_y = pd.merge(delta_df, below_y, how = 'inner', left_index = True, right_index = True)
delta_below_y = delta_below_y[['delta']]

delta_equal_to_y = pd.merge(delta_df, equal_to_y, how = 'inner', left_index = True, right_index = True)
delta_equal_to_y = delta_equal_to_y[['delta']]

delta_higher_than_y = pd.merge(delta_df, higher_than_y, how = 'inner', left_index = True, right_index = True)
delta_higher_than_y = delta_higher_than_y[['delta']]
```

```python
# plot the data
plt.figure()
plt.figure(figsize = (5,6)) # 500px x 600px
ax = plt.gca()

plt.subplots_adjust(left=None, bottom=0.2, right=None, top=None, wspace=0.5, hspace=0.8) # spacing between subplots

# base chart
#plt.bar(df.index, df.mean(axis = 1),
#        width = 1.0,
#        color = 'grey',
#        alpha = 0.2,
#        edgecolor = 'black',
#        linewidth = 1.2,
#        yerr = delta,
#        capsize = 10)

# coloured charts based on condition
plt.bar(below_y.index, below_y.mean(axis = 1),
        width = 1.0,
        color = 'blue',
        alpha = 0.5,
        edgecolor = 'black',
        linewidth = 1.2,
        yerr = delta_below_y['delta'].values,
        capsize = 10)

plt.bar(equal_to_y.index, equal_to_y.mean(axis = 1),
        width = 1.0,
        color = 'green',
        alpha = 0.5,
        edgecolor = 'black',
        linewidth = 1.2,
        yerr = delta_equal_to_y['delta'].values,
        capsize = 10)

plt.bar(higher_than_y.index, higher_than_y.mean(axis = 1),
        width = 1.0,
        color = 'red',
        alpha = 0.5,
        edgecolor = 'black',
        linewidth = 1.2,
        yerr = delta_higher_than_y['delta'].values,
        capsize = 10)

# bar chart tweaking
plt.xticks(df.index.values)

ax.spines['top'].set_visible(False) # don't show any borders around the chart
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.set_title('Mean values and 95% confidence intervals\nfrom normal distribution\nby year') # set title

footnote = '''Note on bar chart colours:
Red: y value < bar_height (i.e. mean)
Green: y value = bar_height
Blue: y value > bar_height
'''
plt.figtext(0.13,0.01, footnote, fontsize=9, va='bottom', ha='left')  # footnote

plt.axhline(y=y_value, color='black', linestyle='-')  # horizontal line

plt.show()
```
